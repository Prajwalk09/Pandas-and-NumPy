## Group By and Aggregation

```python
import pandas as pd
```

```python
data = pd.read_csv(r"C:\Users\prajw\Downloads\Python Files\Jupyter File\Data\Flavors.csv")
data.head()
```

```python
group_by_frame = data.groupby(by='Base Flavor')
```

```python
group_by_frame.mean()
```

```python
data.groupby(by='Base Flavor').mean()
```

```python
data.groupby(by='Base Flavor').count()
```

```python
data.groupby(by='Base Flavor').min()
```

```python
data.groupby(by='Base Flavor').max()
```

```python
data.groupby(by='Base Flavor').sum()
```

### Using the agg() function

Displaying values only in 1 column

```python
data.groupby(by = 'Base Flavor').agg({'Flavor Rating': ['mean']})
```

Displaying values in 1 column but applying multiple aggregate functions

```python
data.groupby(by = 'Base Flavor').agg({'Flavor Rating': ['mean', 'max', 'count', 'sum']})
```

Displaying values in multiple columns but applying multiple aggregate functions for each column

```python
data.groupby(by = 'Base Flavor').agg({'Flavor Rating': ['mean','min','max','sum'], 'Texture Rating':['mean', 'max']})
```

### Grouping based on Multiple columns

```python
data.groupby(by = ['Base Flavor', 'Liked']).mean()
```

```python
data.groupby(by = ['Base Flavor', 'Liked']).sum()
```

```python
data.groupby(by = ['Base Flavor', 'Liked']).agg({'Texture Rating': ['mean', 'sum']})
```

```python
data.groupby(by = ['Base Flavor', 'Liked']).agg({'Texture Rating': ['mean', 'sum'], 
                                                  'Total Rating':  ['mean', 'max']
                                                })
```

### Using the describe() function

```python
data.groupby(by = 'Base Flavor').describe().transpose()
```

```python
data.groupby(by = ['Base Flavor', 'Liked']).describe().transpose()
```
