## Filtering and Ordering DataFrames

```python
import pandas as pd
```

```python
data = pd.read_csv(r"C:\Users\prajw\Downloads\Python Files\Jupyter File\world_population.csv")
data.head()
```

```python
data[data['Rank'] <10]
```

### Using the isin() function

```python
specific_countries = ["Bangladesh", "Brazil", "India", "China"]

data[data['Country'].isin(specific_countries)]
```

### Using the contains() function

#### <font color = 'blue'> Similar to the "like" function in SQL </font>

```python
data[data['Country'].str.contains('United')]
```

Converting the "country" column as index 

```python
new_data = data.set_index('Country')
new_data.head()
```

### Usage of filter() function

```python
new_data.filter(items = ['Continent', 'CCA3', 'Capital'], axis=1)
```

```python
new_data.loc['United States']
```

### Usage of sort_values() function

Sorting Ascendingly

```python
data[data['Rank']<10].sort_values(by = 'Rank',ascending=True)
```

Sorting Descendingly

```python
data[data['Rank']<10].sort_values(by = 'Rank',ascending=False)
```

Sorting based on multiple columns. Both the columns are sorted in ascending order

```python
data[data['Rank']<10].sort_values(by = ['Continent', 'CCA3'], ascending= True)
```

Sorting based on multiple columns. First column is sorted in ascending order and second one is sorted in descending order

```python
data[data['Rank']<10].sort_values(by = ['Continent', 'CCA3'], ascending = [True, False])
```
