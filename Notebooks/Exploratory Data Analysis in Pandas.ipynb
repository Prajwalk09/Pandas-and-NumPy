```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data.corr()
```

```python
plt.rcParams['figure.figsize'] = (14,10)
sns.heatmap(data.corr(), annot = True)
plt.show()
```

```python
data.groupby(by = 'Continent').mean().sort_values(by = '2022 Population', ascending=False)
```

```python
grouped_data = data.groupby(by = 'Continent')['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population'].mean().sort_values(by = '2022 Population', ascending=False).transpose()
grouped_data
```

```python
grouped_data.plot(figsize = (15,5))
plt.show()
```
