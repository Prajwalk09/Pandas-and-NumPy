### Creating index while reading csv file

```python
data = pd.read_csv(r"C:\Users\prajw\Downloads\Python Files\Jupyter File\world_population.csv", index_col='Country')
data.head()
```

### Resetting the index

```python
data.reset_index(inplace=True)
data.head()
```

### Another way to set index

```python
data.set_index('Country', inplace=True)
```

### Searching through an Index

```python
data.loc['India']
```

```python
data.iloc[1]
```

### Multi Indexing

```python
data.reset_index(inplace=True)
```

```python
data.set_index(keys = ['Continent', 'Country'], inplace=True)
```

```python
data.head()
```

```python
data.sort_index(ascending=False)
```

The below code sorts the first index in <strong>descending</strong> order and the second index in <strong>ascending</strong> order

```python
data.sort_index(ascending=[False, True])
```

Searching values based on *first* index

```python
data.loc['Asia']
```

Searching values based on *both* indices

```python
data.loc['Asia'].loc['India']
```
